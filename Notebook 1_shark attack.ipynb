# 3.3 - transformando termos:
aqui estamos transformando tudo que tem a palavra fishing em apenas 'fishing'

```python
##aqui colocamos o nome da coluna e logo em seguida colocamos os termos que queremos encontrar e substituir pela palavra que colocamos logo em seguida, no caso abaixo é fishing
df_act_limpo.loc[df_act_limpo['activity'].str.contains('fishing|Fishing|Spearfishing'),'activity'] = 'fishing'
df_act_limpo['activity'].value_counts()
```

```python
df_act_limpo.loc[df_act_limpo['activity'].str.contains('Swimming|swimming|swimming '), 'activity'] = 'swimming'
df_act_limpo['activity'].value_counts()
```

```python
df_act_limpo.loc[df_act_limpo['activity'].str.contains('Diving|diving|Dived'), 'activity'] = 'diving'
df_act_limpo['activity'].value_counts()
```

```python
df_act_limpo.loc[df_act_limpo['activity'].str.contains('Attempting|attempting'), 'activity'] = 'attempting'
df_act_limpo['activity'].value_counts()
```

```python
df_act_limpo.loc[df_act_limpo['activity'].str.contains('Surfing|surfing'), 'activity'] = 'surfing'
df_act_limpo['activity'].value_counts()
```

```python
df_act_limpo.loc[df_act_limpo['activity'].str.contains('Surf skiing|Surf-skiing'), 'activity'] = 'surf_skiing'
df_act_limpo['activity'].value_counts()
```

# 4.3 - Aqui relacionaremos 'country' x 'area'
É importante analisarmos este dado, pois podemos concatenar com as práticas envolvidas . Aqui daremos um **groupby**, agrupando country e area - utilizaremos a função **size** para saber quantos casos aconteceram ns cidades referidas e também usaremos o **sort_values** para trazer as cidades com maior incidência em primeiro lugar. 

```python
##aqui podemos perceber que a maior quantidade de casos aconteceu na Flórida e em New South Wales
print(df_cou.groupby(["country", "area"]).size().sort_values(ascending=False))
```

```python
#colocamos o reset.index para reverter a lógica do groupby, que transforma as colunas chamadas em index
df_group = df_cou.groupby(["country", "area"]).count()[['case number']].reset_index()
df_group[(df_group['country']== 'USA')|(df_group['country']== 'AUSTRALIA')].sort_values(['country','case number'],ascending = False) 
```
